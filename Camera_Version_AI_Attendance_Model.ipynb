```python
import pandas as pd
import pickle
import base64
import numpy as np

employee_file = '/content/drive/MyDrive/AI Attendance System/System_2/employees.csv'

# Add a new employee to the database
def add_employee_to_csv(employee_id, name, embedding):
    try:
        df = pd.read_csv(employee_file)
    except FileNotFoundError:
        df = pd.DataFrame(columns=["Employee_ID", "Name", "Embedding"])

    # Serialize and encode the embedding
    embedding_serialized = base64.b64encode(pickle.dumps(embedding)).decode('utf-8')

    # Append the new employee data
    new_data = pd.DataFrame([{
        "Employee_ID": employee_id,
        "Name": name,
        "Embedding": embedding_serialized
    }])
    df = pd.concat([df, new_data], ignore_index=True)

    # Save back to CSV
    df.to_csv(employee_file, index=False)
    print(f"Employee {name} added successfully.")

e_id = input()
Name = input()
add_employee_to_csv(e_id, Name, embedding)
```

```python
def delete_employee_from_csv(employee_id):
    # Load the existing database
    df = pd.read_csv(employee_file)

    # Filter out the employee
    df = df[df['Employee_ID'] != employee_id]

    # Save back to Excel
    df.to_csv(employee_file, index=False)
    print(f"Employee with ID {employee_id} deleted successfully.")

delete_e_id = input()
delete_employee_from_csv(delete_e_id)
```

```python
import pandas as pd
import numpy as np
import pickle
import base64
from sklearn.metrics.pairwise import cosine_similarity

def recognize_person_from_csv(face_embedding, cosine_threshold=0.7, euclidean_threshold=20):
    try:
        # Load the employee database
        df = pd.read_csv(employee_file)
    except FileNotFoundError:
        print("Employee database not found.")
        return None, None

    for _, row in df.iterrows():
        # Decode and deserialize the stored embedding from the CSV
        stored_embedding = pickle.loads(base64.b64decode(row['Embedding']))

        # Calculate Euclidean distance
        euclidean_distance = np.linalg.norm(face_embedding - stored_embedding)

        # Calculate cosine similarity
        cosine_sim = cosine_similarity([face_embedding], [stored_embedding])[0][0]

        # Check if either of the conditions is satisfied
        if cosine_sim > cosine_threshold or euclidean_distance < euclidean_threshold:
            return row['Name'], row['Employee_ID']

    return None, None
```

```python
from datetime import datetime

# Path to attendance log Excel file
attendance_file = '/content/drive/MyDrive/AI Attendance System/System_2/attendance_log.xlsx'

# Create an initial attendance log structure (if file doesn't exist)
try:
    pd.read_excel(attendance_file)
except FileNotFoundError:
    columns = ['Employee_ID', 'Name', 'Timestamp']
    pd.DataFrame(columns=columns).to_excel(attendance_file, index=False)

# Record attendance in the Excel file
def record_attendance_in_excel(employee_id, name):
    # Load the existing attendance log
    df = pd.read_excel(attendance_file)

    # Add a new record
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    new_data = {'Employee_ID': employee_id, 'Name': name, 'Timestamp': timestamp}
    df = pd.concat([df, pd.DataFrame([new_data])], ignore_index=True)

    # Save back to Excel
    df.to_excel(attendance_file, index=False)
    print(f"Attendance recorded for {name} at {timestamp}.")
```
